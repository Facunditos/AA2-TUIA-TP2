```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./valores_atributos.csv')
df.head(30)
```

```python
df.describe().style.format(thousands='.',precision=2)
```

```python
plt.figure(figsize=(8,16))
for i,feature in enumerate(df.columns):
    plt.subplot(4,2,i+1)
    sns.boxplot(df[feature])
    plt.ylabel(None)
    plt.title(f'{feature}')
plt.suptitle('Boxplot de los atributos')
plt.tight_layout()
plt.show()
```

```python
df['player_vel'].quantile(q=[0.2,0.4,0.6,0.8])
```

```python
# Distancia tubera prxima
n_bins = 3
primera_fractila = 1 / n_bins
fractilas = [round(primera_fractila,2) * numero for numero in range(1,n_bins+1)]
df['next_pipe_dist_to_player'].quantile(q=fractilas[:-1])
```
